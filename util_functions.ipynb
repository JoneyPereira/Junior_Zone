```python
import pandas as pd
from datetime import datetime, timedelta
import pandas as pd
```

```python
#read csv file

df = pd.read_csv('2023-11-17.csv')
print(df)
```

- Verificando se existe o job_id_list
- Concatenando o dataframe
- Drop duplicados 

```python
import os
cols = ['jobId', 'publishedDate']
df_ids = df[cols]

filename = 'job_id_list.csv'
# verify if the filename file exists or not
if not os.path.exists(filename):
    df_ids.to_csv(filename, index=False)
else:
    csv_file = pd.read_csv(filename)
    csv_file = pd.concat([csv_file, df_ids], ignore_index=True)
    csv_file.drop_duplicates(subset='jobId', inplace=True)
    csv_file.to_csv(filename, index=False)



```

## verificado se existe id

```python
job_id = 6000571

csv_file = pd.read_csv(filename)
if job_id in csv_file['jobId'].values:
    print('Existente')
else:
    print('No existente')
```
